```python
import os
import requests
import urllib.parse

# This is the path where you want to search
path = mypath
files = []
# this is the extension you want to detect
extension = '.csv'

for root, dirs_list, files_list in os.walk(path):
    for file_name in files_list:
        if os.path.splitext(file_name)[-1] == extension:
            file_name_path = os.path.join(root, file_name)
            files.append(file_name_path)
            print(file_name_path)   # This is the full path of the filter file
```

```python
import pandas as pd
import time
add_dict = dict()

previous_add = '' 
temp = []
new_df = pd.DataFrame()
new_df['XCord']=0
new_df['YCord'] = 0
for i1, file in enumerate(files):
    if file in str(temp):
        continue
    temp.append(str(file))
    print("Number:",i1, "of", len(files))
    temp_df = pd.read_csv(file, )
    new_df = temp_df[temp_df['Detected State'].str.contains('Jharkhand', na=False)]
    #print(new_df)
    print(file)
    print(temp)
    for i, j in enumerate(new_df.iterrows()):
        if i%500 == 0 and i>0:
            print('waiting 500 limit')
            #time.sleep(60)
        print(i)
        city = j[1]['Detected City']
        district = j[1]['Detected District']
        state = j[1]['Detected State']

        address = str(city)+' '+str(district) +' '+ str(state)
        address = address.replace("nan","")
        print(address)
        
        if address in add_dict:
            new_df.at[j[0],['XCord']] = add_dict[address][0]
            new_df.at[j[0],['YCord']] = add_dict[address][1]
            print('Found in Dict')
            continue
        if previous_add == address:
            new_df.at[j[0],['XCord']] = prev_x
            new_df.at[j[0],['YCord']] = prev_y
            print("previous")
            print(prev_x)
            print(prev_y)
        else:
            previous_add = address
        
            url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(address) +'?format=json'
            
            response = requests.get(url).json()
            if len(response) != 0:
                print(response[0]["lat"])
                print(response[0]["lon"])
                new_df.at[j[0],['XCord']] = response[0]["lat"]
                new_df.at[j[0],['YCord']] = response[0]["lon"]
                prev_x = response[0]["lat"]
                prev_y = response[0]["lon"]
                add_dict[address] = (response[0]["lat"], response[0]["lon"] )
            else:
                new_df.at[j[0],['XCord']] = 0
                new_df.at[j[0],['YCord']] = 0
                prev_x = 0
                prev_y = 0
                add_dict[address] = (0,0)
        #print(address)
        print()

    coviddataset_df = coviddataset_df.append(new_df, ignore_index = True)
    
    print(len(files))
    print(coviddataset_df.info())
    print('waiting')
    #time.sleep(300)
```
